### Funzione di controllo delle pagine del file excel

Ho scritto una funzione che ha il compito di controllare che i nomi dei comuni delle righe $i$ dellediverse  pagine del foglio excel coincidano.

```python
def controlla_file(df1, df2):
    flag=0
    for i in range(len(df1)):
        if(df1.NOME_COMUNE[i] == df2.NOME_COMUNE[i]):
            flag+=0
        else:
            flag+=1
            
    if(flag==0):
        print('MOLTO BENE, TUTTI I COMUNI CORRISPONDONO NEI DIVERSI FOGLI DI EXCEL, PUOI PROCEDERE!')
        
controlla_file(df_tot_mese, df_tot_sesso)
```

## 1.1 Definisco alcuni DataFrame personalizzati
### 1.1.1 DataFrame comuni

**NOTA:** il numero di decessi totali considerati è, in ciascuna delle colonne, riferito ai mesi marzo, aprile, maggio e giugno (qualsiasi sia l'anno di riferimento). Questo dataframe considera i valori calcolati per ciascun comune. Inoltre, il valore riportato sotto legenda <font color='brown'>[1519]</font> fa riferimento alla dei decessi calcolata sui dati registrati negli anni 2015-1019.

Costruisco un DataFrame, che chiamo <font color='brown'>[mydf]</font>, contenente: 

- 1. **codice regione** <font color='brown'>[mydf.Codice_regione]</font>;
- 2. **regione** <font color='brown'>[mydf.Regione]</font>;
- 3. **provincia** <font color='brown'>[mydf.Provincia]</font>;
- 4. **nome comune** <font color='brown'>[mydf.Comune]</font>;
- 5. **fascia geografica** <font color='brown'>[mydf.Fascia_geografica]</font>: ho suddiviso l'italia in tre fascie, sfruttando i codici numerici delle regioni, forniti dalle tabelle scaricate per l'esame. Questi codici sono numerati in modo tale che le regioni indicate dal range [1,7] siano localizzate a Nord, quelle corrispondenti al range [8,12] siano localizzate al Centro e il range [13,20] si riferisca a Sud e isole;
- 6. **decessi totali 2015-2019** <font color='brown'>[mydf.Totali_1519]</font>;
- 7. **decessi totali anno 2020** <font color='brown'>[mydf.Totali_2020]</font>;
- 8. **decessi totali uomini 2015-2019** <font color='brown'>[mydf.Uomini_1519]</font>;
- 9. **decessi totali donne 2015-2019** <font color='brown'>[mydf.Donne_1519]</font>;
- 10. **decessi totali uomini 2020** <font color='brown'>[mydf.Uomini_2020]</font>;
- 11. **decessi totali donne 2020** <font color='brown'>[mydf.Uomini_2020]</font>;
- 12. **incremento dei decessi totale** <font color='brown'>[mydf.Incremento_totale]</font>: per ottenere questi valori ho utilizzato l'incremento di uomini+donne, riportato dalla tabella apposita, caricata come df_tot_sesso. Ho scelto di normalizzare i valori dell'incremento dividendoli per 100;
- 13. **incremento dei decessi uomini** <font color='brown'>[mydf.Incremento_uomini]</font>;
- 14. **incremento dei decessi donne** <font color='brown'>[mydf.Incremento_donne]</font>;


```python
N = len(df_tot_mese)

morti_per_comune_1519 = list()               #decessi totali per comune media 2015-2019
morti_per_comune_2020  = list()              #decessi totali per comune anno 2020

morti_fem_1519 = list()                      #decessi totali per comune donne 2015-2019
morti_mas_1519 = list()                      #decessi totali per comune uomini 2015-2019

morti_fem_2020 = list()                      #decessi totali per comune donne 2020
morti_mas_2020 = list()                      #decessi totali per comune uomini 2020
    
fascia_geo = list()                          #lista per definire la colonna: Fascia Geografica (Nord, Centro, Sud)

incremento = list()                          #incremento dei decessi totale (Uomini+Donne) per comune
incremento_mas = list()                      #incremento dei decessi uomini
incremento_fem = list()                      #incremento dei decessi donne


for i in range(N):
    
    fem_1519 = df_tot_sesso.iloc[i, 6]
    mas_1519 = df_tot_sesso.iloc[i, 7]
    fem_2020 = df_tot_sesso.iloc[i, 9]
    mas_2020 = df_tot_sesso.iloc[i, 10]
    
    morti_1519  = fem_1519 + mas_1519
    morti_2020  = fem_2020 + mas_2020
    
    morti_per_comune_1519.append(morti_1519)
    morti_per_comune_2020.append(morti_2020)
    
    morti_mas_1519.append(mas_1519)
    morti_fem_1519.append(fem_1519)
    
    morti_fem_2020.append(fem_2020)
    morti_mas_2020.append(mas_2020)


for i in range(N):
    
    if(df_tot_sesso.iloc[i,14]=='-'):                        #sostituisco '-' con '0', 
        df_tot_sesso.iloc[i,14]=0.                           #riportando a 'double' la colonna  

    if(df_tot_sesso.iloc[i,12]=='-'):                        #sostituisco '-' con '0', 
        df_tot_sesso.iloc[i,12]=0.                           #riportando a 'double' la colonna  
        
    if(df_tot_sesso.iloc[i,13]=='-'):                        #sostituisco '-' con '0', 
        df_tot_sesso.iloc[i,13]=0.                           #riportando a 'double' la colonna  
    
    incremento.append(df_tot_sesso.iloc[i, 14]/100.)         #normalizzo le percentuali
    incremento_mas.append(df_tot_sesso.iloc[i, 12]/100.)
    incremento_fem.append(df_tot_sesso.iloc[i, 13]/100.)

    
for i in range(N):
    if(df_tot_mese.iloc[i,0]<=7):
        fascia_geo.append('Nord')
    if(df_tot_mese.iloc[i,0]>7 and df_tot_mese.iloc[i,0]<=12):
        fascia_geo.append('Centro')
    if(df_tot_mese.iloc[i,0]>12):
        fascia_geo.append('Sud')
```

```python
mydf = pd.DataFrame({'Codice_regione': list(df_tot_mese.REG), 'Regione': list(df_tot_mese.NOME_REGIONE), 'Provincia': list(df_tot_mese.NOME_PROVINCIA), 'Comune': list(df_tot_mese.NOME_COMUNE),'Fascia_geografica': fascia_geo, 'Totali_1519': morti_per_comune_1519, 'Totali_2020': morti_per_comune_2020, 'Uomini_1519': morti_mas_1519, 'Donne_1519': morti_fem_1519, 'Uomini_2020': morti_mas_2020, 'Donne_2020': morti_fem_2020, 'Incremento_totale': incremento, 'Incremento_uomini': incremento_mas, 'Incremento_donne': incremento_fem})
```

### 1.1.2 DataFrame Regioni e Province

Ho inoltre deciso di costruire due tabelle, una riferita ai dati raccolti per provincia, l'altra riferita ai dati raccolti per regione.

Il DataFrame **regionale**, che ho chiamato <font color='brown'>[dfreg]</font>, contiene dati in merito a:

- 1. nome regione  <font color='brown'>[dfreg.Regione]</font>;
- 2. decessi totali  <font color='brown'>[dfreg.Totali]</font>;
- 3. decessi nell'anno 2020 <font color='brown'>[dfreg.2020]</font>;
- 4. decessi medi nel quinquennio 2015-2019 <font color='brown'>[dfreg.1519]</font>.

Il DataFrame **provinciale**, che ho chiamato <font color='brown'>[dfprov]</font>, contiene dati in merito a:

- 1. nome provincia <font color='brown'>[dprov.Provincia]</font>;
- 2. nome regione <font color='brown'>[dprov.Regione]</font>;
- 3. fascia geografica <font color='brown'>[dprov.Fascia_geografica]</font>: essendo <font color='brown'>[dfreg]</font> derivata dalla tabella dei comuni, la variabile è la stessa che ho spiegato prima;
- 4. decessi totali per provincia nell'anno 2020 <font color='brown'>[dprov.Decessi_totali_2020]</font>;
- 5. decessi medi totali per provincia nel quinquennio 2015-1019 <font color='brown'>[dprov.Decessi_totali_1519]</font>.

```python
# ------------------------------------ costruzione lista singole regioni --------------------------------

regioni = list()

for i in range(len(df_tot_sesso)):
    count = 0
    for j in range(len(regioni)):
        if(df_tot_sesso.NOME_REGIONE[i]==regioni[j]):
            count= count+1
    if(count==0):
        regioni.append(df_tot_sesso.NOME_REGIONE[i])
        
len(regioni)

#-------------------------------------------- contatori decessi totali per regione ---------------------

def contatore_regionale_1519(nome):
    morti = 0
    for i in range(len(df_tot_sesso)):
        if(mydf.Regione[i]==nome):
            morti+=mydf.Totali_1519[i]
    return morti

def contatore_regionale_2020(nome):
    morti = 0
    for i in range(len(df_tot_sesso)):
        if(mydf.Regione[i]==nome):
            morti+=mydf.Totali_2020[i]  
    return morti

#---------------------------------------------- costruzione liste decessi regionali ----------------------

morti_reg_1519 = list()
morti_reg_2020 = list()

for i in range(len(regioni)):
    morti_reg_1519.append(contatore_regionale_1519(regioni[i]))
    morti_reg_2020.append(contatore_regionale_2020(regioni[i]))
    
#----------------------------------------------- decessi biennio ideale in cui ---------------------------
#------------------------------------------------ un anno è il 2020 --------------------------------------
    
totali_reg = list()
for i in range(len(regioni)):
    totali_reg.append(morti_reg_1519[i]+morti_reg_2020[i])
```

# 3. Analisi della correlazione

Per svolgere l'analisi di correlazione ho deciso inizialmente di visualizzare la clustermap relativa al DataFrame <font color='brown'>mydf</font> (**Figura 4**). In seguito, ho rappresentato in un grafico a dispersione i dati relativi ai decessi totali registrati per provincia, mantenendo distinta la classificazione per fascia geografica e rappresentando le distribuzioni marginalizzate per gli anni 2020 e media 15-19 (**Figura 5**).

## 3.1  Analisi della matrice di correlazione

```python
sns.clustermap(mydf.corr(), annot=True, figsize=(15,15), vmin=-1, vmax=+1, cmap='coolwarm', linewidth=.4)

plt.tight_layout()
plt.show()
```

**Figura 4:**

- **Commento 1:** la clustermap rappresenta la matrice di correlazione dei dati utilizzati. Sono evidenziati alcuni raggruppamenti nella correlazione delle variabili: il primo blocco, ben evidente in rosso, corrisponde a tutti i dati che, in qualche maniera, si riferiscono al numero totale di decessi (indipendentemente dall'anno). In questo gruppo si registra una chiara correlazione positiva tra le variabili, indice del fatto che una sola di queste potrebbe bastare a raccogliere il contenuto informativo del campione in merito al numero di decessi.

- **Commento 2:** si nota inoltre un secondo blocco che indica correlazione positiva analizzando le variabili relative all'incremento della mortalità. In particolare, notiamo bassa ($+0.18$) correlazione tra incrementi di mortalità per generi diversi ma correlazione simile tra l'incremento totale e l'incremento di ciascun genere: $\rho_{tot, mas}=+0.69, \rho_{tot,fem}=+0.66$. - **Commento 3:** un terzo blocco evidenzia anticorrelazione tra il codice numerico identificativo delle regioni e l'incremento dei decessi. # 4. Stima puntuale di parametri e test di bontà dei fit

L'idea sviluppata nella prossima parte del notebook è la seguente: volendo ulteriormente analizzare le differenze in merito all'incremento della mortalità registrato nell'anno della pandemia in funzione della fascia geografica, ho deciso di effettuare un fit di Maximum-Likelihood utilizzando una distribuzione Gamma per i dati relativi ai comuni e suddivisi in nord, centro e sud (comprese le isole). La distribuzione Gamma dipende anche da un parametro di forma, il quale può evidenziare eventuali asimmetrie nell'istogramma dell'incremento dei decessi (che immagino a destra dello zero) e, tramite MLE, voglio provare a quantificare la differente reazione al virus delle fasce geografiche confrontando i parametri di forma delle tre distribuzioni. 

Aspettandomi infine che in assenza di epidemia la distribuzione degli incrementi sia simmetrica rispetto allo zero, ho studiato nel dettaglio la tabella riguardante i dati della fascia geografica più a sud, che dovrebbe essere stata la meno colpita dal virus.

### Divisione dei dataset in nord, centro e sud e isole

```python
# la scelta: pongo uguali a zero gli incrementi negativi
my_df_nord = mydf[mydf.Fascia_geografica=='Nord']
my_df_sud = mydf[mydf.Fascia_geografica=='Sud']
my_df_centro = mydf[mydf.Fascia_geografica=='Centro']

mydf_inc = mydf[mydf.Incremento_totale>0]

print('Numero comuni analizzati a nord, centro e sud:  ', len(my_df_nord), ", ", len(my_df_centro), ", ", len(my_df_sud))
```

## 4.1) Stime di maximum-likelihood

### Gamma fit sui tre campioni relativi alle fasce geografiche


```python
sud_fit_g = sc.gamma.fit(my_df_sud.Incremento_totale)                       # MLE Gamma fascia sud

cen_fit_g = sc.gamma.fit(my_df_centro.Incremento_totale)                    # MLE Gamma fascia centro
nor_fit_g = sc.gamma.fit(my_df_nord.Incremento_totale)                      # MLE Gamma fascia nord


zz = np.linspace(-1.5,4, 10000)                                             # linspace per rappresentare le pdf


sud_pdf_g = sc.gamma.pdf(zz, sud_fit_g[0], sud_fit_g[1], sud_fit_g[2])      # pdf Gamma sud
cen_pdf_g = sc.gamma.pdf(zz, cen_fit_g[0], cen_fit_g[1], cen_fit_g[2])      # pdf Gamma centro
nor_pdf_g = sc.gamma.pdf(zz, nor_fit_g[0], nor_fit_g[1], nor_fit_g[2])      # pdf Gamma nord


print('                                                                                                       ')
print('*******************************************************************************************************')
print('**                     MLE GAMMA A CONFRONTO PER FASCE GEOGRAFICHE                                   **')
print('*******************************************************************************************************')
print('                                                                                                       ')
print('STIMATORI DI ML PER MODELLO GAMMA E CAMPIONE NORD   :', nor_fit_g)   
print('STIMATORI DI ML PER MODELLO GAMMA E CAMPIONE CENTRO :', cen_fit_g)
print('STIMATORI DI ML PER MODELLO GAMMA E CAMPIONE SUD    :', sud_fit_g)
```

### Fit di ML del campione relativo alla fascia: sud e isole; Gaussiana, Cauchy, Gamma e Laplace a confronto,

```python
sud_fit = sc.norm.fit(my_df_sud.Incremento_totale)                          # MLE Gauss fascia sud
sud_fit_l = sc.laplace.fit(my_df_sud.Incremento_totale)                     # MLE Laplace fascia sud
sud_fit_c = sc.cauchy.fit(my_df_sud.Incremento_totale)                      # MLE Cauchy fascia sud

sud_pdf = sc.norm.pdf(zz, sud_fit[0],sud_fit[1])                            # pdf Gauss sud
sud_pdf_l = sc.laplace.pdf(zz, sud_fit_l[0],sud_fit_l[1])                   # pdf Laplace sud
sud_pdf_c = sc.cauchy.pdf(zz, sud_fit_c[0], sud_fit_c[1])                   # pdf Cauchy sud


sud_cdf = sc.norm.cdf(zz, sud_fit[0],sud_fit[1])                            # cdf Gauss sud
sud_cdf_g = sc.gamma.cdf(zz, sud_fit_g[0], sud_fit_g[1], sud_fit_g[2])      # cdf Gamma sud
sud_cdf_l = sc.laplace.cdf(zz, sud_fit_l[0],sud_fit_l[1])                   # cdf Laplace sud
sud_cdf_c = sc.cauchy.cdf(zz, sud_fit_c[0], sud_fit_c[1])                   # cdf Cauchy sud


print('                                                                                                       ')
print('*******************************************************************************************************')
print('**                MLE GAUSS, GAMMA, LAPLACE E CAUCHY FASCIA SUD                                      **')
print('*******************************************************************************************************')
print('                                                                                                       ')

print('STIMATORI DI ML PER MODELLO GAUSS E CAMPIONE SUD    :', sud_fit)   
print('STIMATORI DI ML PER MODELLO LAPLACE E CAMPIONE SUD  :', sud_fit_l)
print('STIMATORI DI ML PER MODELLO GAMMA E CAMPIONE SUD    :', sud_fit_g)
print('STIMATORI DI ML PER MODELLO CAUCHY E CAMPIONE SUD   :', sud_fit_c)
```

### Rappresentazione grafica dei Fit 

```python
plt.figure(figsize=(15,10))

plt.subplot(2,7,(1,7))
plt.title("(A) Istogrammi dell'incremento di mortalità e Gamma Fit, confronto per fascia geografica")
plt.hist(my_df_nord.Incremento_totale, bins=30, alpha=0.4, density=True, color='blue', label='Nord', range=(-2,6))
plt.hist(my_df_nord.Incremento_totale, bins=30, alpha=1, histtype='step', linewidth=2, density=True, color='blue', range=(-2,6))
plt.plot(zz, nor_pdf_g, label='ML Fit Gamma NOR', color='royalblue', linewidth=3, alpha=0.9)

plt.hist(my_df_centro.Incremento_totale, bins=30, alpha=0.4, density=True, color='forestgreen', label='Centro')
plt.hist(my_df_centro.Incremento_totale, bins=30, alpha=1, histtype='step', linewidth=2, density=True, color='forestgreen')
plt.plot(zz, cen_pdf_g, label='ML Fit Gamma CEN', color='limegreen', linewidth=3, alpha=0.9)

plt.hist(my_df_sud.Incremento_totale, bins=30, alpha=0.4, density=True, color='red', label='Sud')
plt.hist(my_df_sud.Incremento_totale, bins=30, alpha=0.7, histtype='step', linewidth=2, density=True, color='red')
plt.plot(zz, sud_pdf_g, label='ML Fit Gamma SUD', color='orangered', linewidth=3, alpha=0.9)

plt.xlabel('Incremento dei decessi')
plt.ylabel('N normalizzato')


plt.legend()
plt.subplot(2,7,(8,10))
plt.title(' (B) Incremento al sud, fit p.d.f.')
plt.hist(my_df_sud.Incremento_totale, bins=30, alpha=0.5, density=True, color='red', label='Dati')
plt.plot(zz, sud_pdf, label='ML Fit Gauss', color='royalblue', linewidth=3, alpha=0.9)
plt.plot(zz, sud_pdf_g, label='ML Fit Gamma', color='limegreen', linewidth=3, alpha=0.9)
plt.plot(zz, sud_pdf_c, label='ML Fit Cauchy', color='black', linewidth=3, alpha=0.9)
plt.plot(zz, sud_pdf_l, label='ML Fit Laplace', color='gold', linewidth=3, alpha=0.9)

plt.xlim(-2,2)
plt.xlabel('Incremento dei decessi')
plt.ylabel('N normalizzato')
plt.legend()


plt.subplot(2,7,(11,14))
plt.title('(C) Incremento al sud, fit c.d.f.')
plt.hist(my_df_sud.Incremento_totale, bins=30, alpha=0.5, density=True, color='red', cumulative=True, label='Dati')
plt.plot(zz, sud_cdf_g, label='ML Fit Gamma', color='limegreen', linewidth=3, alpha=0.9)
plt.plot(zz, sud_cdf, label='ML Fit Gauss', color='royalblue', linewidth=3, alpha=0.9)
plt.plot(zz, sud_cdf_c, label='ML Fit Cauchy', color='black', linewidth=3, alpha=0.7)
plt.plot(zz, sud_cdf_l, label='ML Fit Laplace', color='gold', linewidth=3, alpha=0.9)

plt.xlim(-1.5,2)
plt.ylabel('N normalizzato')
plt.xlabel('Incremento dei decessi')

plt.legend()

plt.tight_layout()


plt.show()
```

**Figura 6:** in figura 6.A sono rappresentati i tre istogrammi normalizzati dei dati relativi all'incremento dei decessi suddivisi per fascia geografica. Per ciascun campione ho effettuato un fit di maximum-likelihood avendo fissato come modello una distribuzione Gamma. Già graficamente è evidente che spostandosi verso nord lungo la penisola la curva risulta "sbilanciata" verso destra, ossia che la coda riguardante l'incremento positivo dei decessi sottenda più dati. Questo si spiega ricordando che il virus ha colpito in modo più aggressivo il nord, per motivi di contagio zero e di densità demografica. A supporto dell'evidenza grafica, ho stampato i valori delle stime puntuali di parametro effettuate tramite scipy: il primo parametro, ossia il primo elemento dei vettori corrispondenti ai fit, contiene le informazioni riguardanti la forma della distribuzione. I valori, che da nord a sud sono: 4.57, 12.82, 21.31, sottolineano come l'asimmetria del campione riferito al nord sia maggiore rispetto a quella del campione centro, a sua volta maggiore di quella riferita al sud; Si noti che se il valore di questo parametro tende ad uno la distribuzione Gamma tende ad una distribuzione esponenziale. Il terzo parametro, corrispondente ad un fattore di scala, indica inoltre che nel nord vi sia una dispersione dei dati relativi all'incremento maggiore che nelle rimanenti fasce geografiche.

Ipotizzando che ciò sia dovuto al fatto che il sud sia stato colpito meno gravemente dalla pandemia (forse per motivi demografici, forse per minore diffusione del virus, forse per cause climatiche), ho deciso di studiare più nel dettaglio il campione relativo a questa fascia geografica, ritenendolo più regolare. Al fine di poter trarre conclusioni o rigettare alcuni dei modelli sperimentati, nel paragrafo successivo ho riportato una serie di test statistici, basati sulla statistica di Kolmogorov-Smirnov.

### Commento sul picco localizzato in $X=-1$

La scheda riepilogativa dei dati spiega che i valori della colonna che ho analizzato sono stati ottenuti tramite l'espressione:

$$ X = \frac{L-I}{I}\cdot 100,$$

ossia sottraendo ai decessi nel 2020 ($L$) il valore contenuto nella colonna riferita alla media negli anni 2015-2019 ($I$). Il caso di valore $-100$  (corrispondente al mio $-1$), si ottiene se $L=0$, ovvero se nei tre mesi del 2020 sono stati registrati in totale 0 decessi. Cosa probabile per comuni piccoli e isolati, probabilmente non colpiti dalla pandemia.

Tramite una funzione di controllo ho conteggiato il numero di osservazioni tali che $X=-1$, riscontrando $466$ dati su un totale di $7357$. Questo valore, a mio parere, modifica in modo consistente la distribuzione: tutti i numeri a virgola mobile rappresentanti i decessi nel quinquennio 2015-2019 e descriventi piccole fasce di incremento di mortalità (-1, +1), sono classificate come $-100$ dalla formula precedentemente esposta se (e non è improbabile per i piccoli comuni) il numero di decessi negli stessi mesi del 2020 è stato pari a zero.

## 4.2) K-S Test tra campioni e modelli ottenuti tramite i fit di ML

### Distribuzione Gamma e fasce geografiche

```python
print('NORD   :', sc.kstest(my_df_nord.Incremento_totale, 'gamma', args=(nor_fit_g[0], nor_fit_g[1], nor_fit_g[2])))
print('CENTRO :', sc.kstest(my_df_centro.Incremento_totale, 'gamma', args=(cen_fit_g[0], cen_fit_g[1], cen_fit_g[2])))
print('SUD    :', sc.kstest(my_df_sud.Incremento_totale, 'gamma', args=(sud_fit_g[0], sud_fit_g[1], sud_fit_g[2])))
```

**Commento:** i $p$-value relativi al fit basato sulla distribuzione Gamma sono $p_i < 10^{-7}$. Ritengo che tale valore sia in parte dovuto all'oggettiva incompatibilità del campione con un modello di questo tipo e in parte dovuto alle irregolarità che ho riscontrato nel campione (ad esempio la sovrastima dell'esatto valore $-1$). Per queste ragioni sento di poter rigettare l'ipotesi che sia una Gamma a descrivere il campione relativo al Nord ma di non poter fare altrettanto per i due campioni rimanenti.

### Modelli Gamma, Gauss, Laplace e Cauchy e campione relativo alla fascia sud

```python
print('LAPLACE  :', sc.kstest(my_df_sud.Incremento_totale, 'laplace', args=(sud_fit_l[0], sud_fit_l[1]), N=len(my_df_sud)))
print('GAUSS    :', sc.kstest(my_df_sud.Incremento_totale, 'norm', args=(sud_fit[0], sud_fit_l[1]), N=len(my_df_sud)))
print('GAMMA    :', sc.kstest(my_df_sud.Incremento_totale, 'gamma', args=(sud_fit_g[0], sud_fit_g[1], sud_fit_g[2])))
print('CAUCHY   :', sc.kstest(my_df_sud.Incremento_totale, 'cauchy', args=(sud_fit_c[0], sud_fit_c[1]), N=len(my_df_sud)))
```

**Commento 1:** a seguito dell'utilizzo della statistica di K-S, è evidente che l'ipotesi meno additata di rigetto sia quella di una distribuzione di Laplace. La distribuzione di Laplace è simmetrica, definita per casi e presenta un comportamento simile ad un'esponenziale per entrambe le code. Riterrei interessante approfondire l'analisi di compatibilità tra un campione "standard" (il sud è infatti la fascia geografica ad aver subito meno la pandemia) di incremento della mortalità e la distribuzione di Laplace.

**Commento 2:** in accordo con quanto già detto in merito al picco anomalo di incrementi corrispondente a $X=-1$, non ritengo sia opportuno scartare alcuna delle distribuzioni analizzate. Un'analisi più approfondita, sulla base di dati più numerosi e regolari, porterebbe certamente a compiere una scelta più decisa.

### Discriminazione per regione

Sfruttando la statistica di Kolmogorov-Smirnov, ho deciso di calcolare i $p$-value per campioni di mortalità totale a confronto regione per regione. Ho ripetuto il calcolo mettendo a confronto anche i campioni regionali di incremento totale.

#### Funzione ricerca della minima incompatibilità, esempio con regione Marche

```python
def test_reg_reg(nome):
    
    tester = 0
    df_rif = mydf[mydf.Regione==nome]
    
    for i in range(len(regioni)):
        df_reg = mydf[mydf.Regione==regioni[i]]
        ktest = sc.ks_2samp(df_rif.Totali_2020, df_reg.Totali_2020)
       # print('KS-Test tra ', nome, ' e ', regioni[i], ': ', ktest[1])
      #  print('                  ')
        if(ktest[1]!=1):
            if(tester<ktest[1]):
                index = i
            tester=max(tester, ktest[1])

    print('  ')
    print('  ')
    print('MIGLIOR P-VALUE: ', tester, ',  REGISTRATO TRA: ', nome, ' e ', regioni[index])
    print('+ + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + + +')
    print('Il campione riguardante i defunti totali nei mesi gennaio-giugno del  2020 della regione che risulta più compatibile con quello di ', nome, 'riguarda la regione ', regioni[index], '!!!')
           
```

```python
test_reg_reg('Marche')
```

#### Calcolo e rappresentazione grafica dei p-value

```python
PV_decessi = np.zeros((20,20))
PV_increm  = np.zeros((20,20)) 


for i in range(20):
    df_reg_host = mydf[mydf.Regione==regioni[i]]
    
    for j in range(20):
        df_reg = mydf[mydf.Regione==regioni[j]]
    
        PV_decessi[i][j]=sc.ks_2samp(df_reg_host.Totali_2020, df_reg.Totali_2020)[1]
        PV_increm[i][j]=sc.ks_2samp(df_reg_host.Incremento_totale, df_reg.Incremento_totale)[1]
```

```python
L = np.linspace(0.5,19.5,21)


plt.figure(figsize=(18,9))

plt.subplot(1,2,1)
plt.title('p-values morti totali/comune')
plt.pcolormesh(regioni, regioni, PV_decessi, cmap='OrRd', vmin=0, vmax=+1, alpha=0.7)
#ax = sns.heatmap(PV, vmin=0, vmax=1,alpha=0.75, annot=True,linewidth=1, cmap='RdYlGn')
plt.yticks(L, regioni)
plt.xticks(L, regioni, rotation=90)
plt.tight_layout()


plt.subplot(1,2,2)
plt.title('p-values incremento/comune')
plt.pcolormesh(regioni, regioni, PV_increm, cmap='OrRd', vmin=0, vmax=+1, alpha=0.7)
plt.yticks(L, regioni)
plt.xticks(L, regioni, rotation=90)
plt.tight_layout()


plt.tight_layout()

plt.show()
```

**Figura 7:** In figura ho voluto rappresentare graficamente i risultati ottenuti nel calcolo del $p$-value per il confronto tra due campioni regionali e per due diverse variabili:

- decessi totali per comune;
- incremento dei decessi totali per comune.

Per quanto il $p$-value sia un risultato inconcludente dal punto di vista della teoria, ho voluto evidenziare eventuali relazioni tra regioni su cui provare ad approfondire l'analisi. Se nel primo dei casi non ho notato nessun comportamento di cluster  a mio parere rilevante, il secondo grafico evidenzia come al nord vada rigettata l'ipotesi che il modello descrivente l'incremento dei decessi sia generalizzabile a regioni diverse. Al contrario, al sud, seppur con risultati instabili, potrebbe essere interessante approfondire l'idea opposta, ossia che i campioni relativi alle regioni meridionali possano essere statisticamente paragonabili.

Ciononostante, i risultati ottenuti nell'ambito dell'analisi di correlazione e del test di ipotesi indicano, al contrario, che è possibile si registrino comportamenti simili tra i due campioni statistici. 

In conclusione, l'analisi delle differenze per fascia geografica e genere ha condotto ai risultati che ho presentato. Essendo costruiti sul confronto di due campioni di variabili diverse (una continua e una discreta), ritengo sia opportuno approfondire questa indagine una volta raccolti i campioni sufficienti per poter confrontare variabili dello stesso tipo, ad esempio ricavando una media mensile dei decessi in periodo di epidemia o calcolandola considerando il biennio 2020-2021. In questo modo, si potrebbero trarre conclusioni più forti in merito all'incremento della mortalità nazionale in questo periodo di epidemia sotto tali discriminazioni di classe.
